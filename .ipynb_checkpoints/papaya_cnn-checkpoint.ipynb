# Diagnosis of Fruit Diseases using image processing and machine learning algorithms

This is a convolutional neural network for detect the main diseas in papaya tree. we have used Tensorflow and Numpy use as a main classes and we have tried to develop a model with minimum number of data points.


## Calling Main classes

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

## Upload the Data using path and categories

mainly, we categorised the data according to the diseases that we want to analyse. And we genarated the sample data point in the dataset for checking accuracy of the function. 

```python
DataDir="papaya"
CATEGORIES=["anthracnose","Phytophthora","ring_spot"]


for category in CATEGORIES:
    path=os.path.join(DataDir,category)
    for img in os.listdir(path):
        img_array= cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array,cmap="gray")
        plt.show()
        break 
    break    
```

```python
print(img_array.shape)
```

## Normalize the dataset

According to above <code> img_array.shape </code>, we know that the data points are in various shapes. then we have normalized the dataset.

```python
IMG_SIZE= 80

new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
plt.imshow(new_array,cmap="gray")
plt.show()
```

## Build the training data set

```python
training_data=[]

def create_td():
    for category in CATEGORIES:
        path=os.path.join(DataDir,category)
        class_num=CATEGORIES.index(category)
        for img in os.listdir(path):
            try:
                img_array= cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
create_td()          
```

```python
print(len(training_data))
```

```python
import random

random.shuffle(training_data)

for sample in training_data[:10]:
    print(sample[1])
```

```python
x=[]
y=[]

for features,label in training_data:
    x.append(features)
    y.append(label)
    
x=np.array(x).reshape(-1,IMG_SIZE,IMG_SIZE,1)
```

```python
import pickle

pickle_out= open("x.pickle","wb")
pickle.dump(x,pickle_out)
pickle_out.close()

pickle_out= open("y.pickle","wb")
pickle.dump(y,pickle_out)
pickle_out.close()

pickle_in= open("x.pickle","rb")
x=pickle.load(pickle_in)
```

```python
x.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D,MaxPooling2D

x=pickle.load(open("x.pickle","rb"))
y=pickle.load(open("y.pickle","rb"))

x = np.array(x)
y = np.array(y)

x=x/255.0
```

## Develop the CNN and Train

We have used RELu as a hiddle layers activation function and used Sigmoid as a output layer activation function. we have used 10 bagging steps.

```python
model=Sequential()
model.add(Conv2D(128,(3,3), input_shape=x.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))

model.add(Dense(1))
model.add(Activation("sigmoid"))

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])

model.fit(x,y,batch_size=10,epochs=10,validation_split=0.1)
```

## Model Evaluation

We have went through training data, because there are no testing data set in here. And expect to improve data set in future. According to the above model evaluation, The model does not well trained because of minority of training data.

```python
y_train= model.predict(x)

y_train
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y,y_train)
```

This accuracy score is not acceptable. we can improve this increasing number of data points and adding testing data set. And also we can use the GAN as a other option.So,most of time output results will be failier.
